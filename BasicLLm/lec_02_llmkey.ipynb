LLM = Large Language Model

It is a very smart computer program that can Read and understand human language (English, Hindi, etc.)
Write answers that sound like a real person and Chat with you, answer questions, write stories, explain things, etc.

It learns patterns from huge amounts of text (books, websites, articles—like the entire internet!).It guesses the next word in a sentence based on what came before.
Example: If I say, "The sky is ___," it might predict "blue" because that's a common pattern in English.

You write: "What is the capital of India?"
The LLM answers: "The capital of India is New Delhi."
But what is actually happening inside the computer right now?
Here are the real steps (like a factory line):
Step 1: Your words → broken into small pieces (Tokens)
Your sentence is chopped into little bits.
Example:
"What" + " is" + " the" + " capital" + " of" + " France" + "?"
These pieces are called tokens.
(Some tokens are whole words, some are parts of words like "ing" or "un-".)

Computer can't read words — it only understands numbers.
So every token becomes a long list of numbers (like a secret code for that word).
Example:

"What" → [0.2, -1.5, 0.8, … 512 numbers]
"is"   → [0.1, 0.9, -0.3, … 512 numbers]

This list of numbers is called an embedding.

```python
import ollama

response = ollama.chat(
    model='llama3.2',
    messages=[
        
        {'role': 'user', 'content': 'Explain quantum entanglement like I am 12 years old'},
    ]
)
print(response['message']['content'])
```

Streaming response if you want from ollama 

```python
from ollama import chat

stream = chat(
    model='llama3.2:latest',
    messages=[
        {'role': 'user', 'content': 'Explain quantum entanglement like I am 12 years old'}],
    stream=True
)

for chunk in stream:
    print(chunk['message']['content'], end='', flush=True)
```

```python
from openai import OpenAI

client = OpenAI(
    base_url='http://localhost:11434/v1/',   # ← This is the magic
    api_key='ollama'                         # Required by the library, but ignored by Ollama
)

response = client.chat.completions.create(
    model='llama3.2:latest',                 # Use any model you pulled with Ollama
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Explain agentic AI in simple terms."}
    ],
    temperature=0.7,
    max_tokens=500,
)

print(response.choices[0].message.content)
```

Stream version of OpenAi library

```python
from openai import OpenAI

client = OpenAI(
    base_url='http://localhost:11434/v1/',   # ← This is the magic
    api_key='ollama'                         # Required by the library, but ignored by Ollama
)
stream = client.chat.completions.create(
    model='llama3.2:latest',
messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Explain agentic AI in simple terms."}
    ],
            stream=True
)

for chunk in stream:
    if chunk.choices[0].delta.content is not None:
        print(chunk.choices[0].delta.content, end="", flush=True)
```
