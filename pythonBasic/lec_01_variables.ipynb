In Python, you cannot define a variable with a specific type "manually" the way you do in languages like C, Java, C#, TypeScript, etc.Python uses dynamic typing by default — the type is determined automatically at runtime when you assign a value.
However, there are several modern ways to  declare  the expected type of a variable. These are called type hints (introduced in Python 3.5, greatly improved in 3.9–3.12).


```python
# These are NOT enforced at runtime — they are for tools & documentation
age: int = 25
name: str = "Vikash"
height: float = 5.7
is_student: bool = True
print(age)
```

Python is an object-oriented language, so every value (number, text, list, etc.) is actually an object in memory.
We usually measure the object size using sys.getsizeof().It tells you approximately how much memory a single object is using right now.

```python


import sys  # We need this module to use getsizeof()

name = "Vikash"                    # string (str)
age = 25                           # integer (int)
height = 5.7                       # floating point (float)
is_learning = True                 # boolean (bool)
city = "Delhi"                     # another string
nothing = None                     # NoneType
print("Variable          Value                  Type              Size (bytes)")
print("-" * 70)
variables = [
    ("name", name),
    ("age", age),
    ("height", height),
    ("is_learning", is_learning),
    ("city", city),
    ("nothing", nothing)
]

for var_name, value in variables:
    size = sys.getsizeof(value)
    type_name = type(value).__name__
    print(f"{var_name:<15} {str(value):<22} {type_name:<17} {size:>3}")
```

Type conversion means changing one data type to another.
Python does this in two ways:

Implicit (automatic — Python does it for you sometimes).
Explicit (manual — you tell Python how, using functions like int(), str()).

Implicit conversion  happens: Small to big type (int → float, but not str → int).

```python
a = 10          # int
b = 3.14        # float
c = a + b       # Python auto-changes int to float
print(c)        # 13.14  (type: float)
print(type(c))  # <class 'float'>
```

These are some functions you actually call when you want to manually change or explicit convert one data type to another (explicit conversion) like int()/str() etc

```python
int("25")          # → 25
float(25)          # → 25.0
bool(1)            # → True
```

A reference is simply a name (like a label or pointer) that points to an object in memory.When you write x = 100 Python creates an integer object with value 100 in memory (or reuses an existing one).The name x becomes a reference to that object.x does not contain the number 100 itself — it just points to the place in memory where 100 lives.

```python
# 1. Numbers (immutable)
x = 10
y = x
x = 20          # creates NEW int object
print("here is X ",x)        # still 10
print("here is Y ",y)        # still 10

```
