Conditions  allow the program to execute certain code only if a condition is true.Python mainly uses if, elif, and else for conditional statements.

If condition Executes code only if the condition is True.

```python
age = 18

if age >= 18:
    print("You are eligible to vote")
```

2. if-else StatementUsed when you want one block if the condition is true and another if it is false.

```python
number = 5

if number % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

3. if-elif-else Statement is Used when there are multiple conditions.

```python
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 70:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

Python uses indentation (spaces or tabs at the beginning of lines) to define which lines belong to the same block of code (if, for, while, def, class, with, try, etc.).This is very different from languages like C, Java, JavaScript → they use { } curly braces.
Python says: "readable code is good code — so let's force people to write readable code".

 Rules of Indentation (must follow)

After a line ending with : → the next line(s) must be indented (start a new block)
Recommended: 4 spaces per level (official PEP 8 style)


```python
if age >= 18:
    print("Adult")           # 4 spaces
    print("Can vote")        # same 4 spaces
else:
    print("Minor")           # 4 spaces
    print("Cannot vote")     # same 4 spaces

print("This runs always")    # no indentation → outside the if-else
```

Logical operators (combine conditions) with if conditions


```python
age = 17
has_voter_id = True

if age >= 18 and has_voter_id:
    print("Can vote")
else:
    print("Cannot vote yet")
```

Quick one-liner style (shorthand if-else) Python

```python
age=20
if age >= 18:
    status = "Adult"
else:
    status = "Minor"

# Short way (ternary operator)
status = "Adult" if age >= 18 else "Minor"
print(status)
```

Nested conditional expressions (chained ternaries)
Python.Python reads this expression from left to right, but the else part is attached to the closest if.

```python
x=50
"high" if x > 100 else "medium" if x > 50 else "low"
```

This is exactly the same as above code

```python
x=10
"high" if x > 100 else ("medium" if x > 50 else "low")
```

Nested (readable version – avoid deep nesting)

```python
score=75
grade = (
    "A" if score >= 90 else
    "B" if score >= 80 else
    "C" if score >= 70 else
    "D" if score >= 60 else
    "F"
)
print(grade)
```

Loops in Python are used to repeat a block of code multiple times.
Python has mainly two types of loops:

for loop → best when you know how many times to repeat (or when iterating over something)

while loop → best when you don't know in advance how many times it will run (condition-based)

```python


print("looping by range(5) function")
for i in range(5):          # 0,1,2,3,4
    print(i)
print("looping by range(1,6) function")

for i in range(1, 6):       # 1 to 5
    print(i)
print("looping by range(1,10,2) function")
for i in range(1, 10, 2):   #/*3,5,7,9 (start, stop, step)
    print(i)


print("looping nested")
for i in range(1, 6):
    for j in range(i):
        print("*", end=" ")
    print()   # new line
```

while loop → best when you don't know in advance how many times it will run (condition-based)

```python
count = 1

while count <= 5:
    print(count)
    count += 1          # very important – otherwise infinite loop!
```

break in Python is a loop control statement that immediately exits / terminates the nearest enclosing loop (for or while) — no matter how many iterations were left.

```python
for i in range(1, 11):
    if i == 6:
        break          # ← stops the loop right here
    print(i)
```

What continue does

Skips the rest of the current iteration
Immediately jumps to the next iteration of the loop
Does not exit the loop (unlike break)

```python
for i in range(1, 11):
    if i % 2 == 0:
        continue
    print(i)      
```
