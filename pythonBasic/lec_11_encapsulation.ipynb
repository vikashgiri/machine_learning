```python
class Person:
    def __init__(self, name, age):
        self.name = name          # public → anyone can access/change
        self._age  = age          # protected → convention: don't touch directly

    def birthday(self):
        self._age += 1
        print(f"Happy birthday! Now {self.name} is {self._age} years old.")

# Usage
p = Person("Vikash", 25)
print(p.name)       # OK → Vikash
p.name = "Vik"      # OK (but maybe not good idea)
print(p._age)       # 25 → technically works, but bad style
p.birthday()        # Good way — uses method
```

_age is protected by convention — programmers agree: "don't touch _age directly, use methods like birthday()"

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance       # ← private (name mangled)

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Added ₹{amount}. Balance now: ₹{self.__balance}")
        else:
            print("Amount must be positive!")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ₹{amount}. Balance now: ₹{self.__balance}")
        else:
            print("Not enough money or invalid amount!")

    def check_balance(self):
        return self.__balance

# Usage — good way
acc = BankAccount("Vikash", 10000)
acc.deposit(5000)          # Works
acc.withdraw(2000)         # Works
print(acc.check_balance()) # 13000

# Bad / blocked ways
# print(acc.__balance)     → AttributeError!
# acc.__balance = -100000  → creates NEW variable — doesn't change real balance
```

What happens behind the scenes?
Python changes __balance → _BankAccount__balance (name mangling)
→ Very hard to access accidentally from outside.
Real protection?
Not 100% (you can still do acc._BankAccount__balance = 999) — but it strongly discourages misuse.

In Python, getters and setters are most commonly created using the @property decorator.
This is the modern, clean, and Pythonic way (called properties).

```python
class Person:
    def __init__(self, name, age):
        self._name = name
        self._age = age

    # Getter for name
    @property
    def name(self):
        return self._name

    # Setter for name
    @name.setter
    def name(self, value):
        if not value.strip():
            raise ValueError("Name cannot be empty")
        self._name = value.title()   # auto-capitalize

    # Getter only (read-only)
    @property
    def age(self):
        return self._age

    # Optional: setter with validation
    @age.setter
    def age(self, value):
        if value < 0:
            raise ValueError("Age cannot be negative")
        self._age = value

# Usage — looks like normal attributes!
p = Person("vikash", 25)

print(p.name)       # vikash          ← calls getter
p.name = "Aman"     # calls setter    → stored as "Aman"
print(p.name)       # Aman

print(p.age)        # 25
# p.age = -5        # → ValueError (because no setter? Wait — we added one)
p.age = 26
print(p.age)        # 26
```
