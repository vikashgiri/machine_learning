Syntex to create a dictonary

```python
# 1. Create / update
scores = {}
scores["Vikash"] = 85
scores["Priya"]  = 92
scores["Rahul"]  = 78

# or at once
scores = {"Vikash": 85, "Priya": 92, "Rahul": 78}

# Update existing
scores["Vikash"] = 88
scores.update({"Ananya": 95, "Rahul": 80})

print(scores)
# {'Vikash': 88, 'Priya': 92, 'Rahul': 80, 'Ananya': 95}
```

```python
# 2. Reading values (safe way)
print(scores["Priya"])          # 92
print(scores.get("Rohan"))      # None (no error)
print(scores.get("Rohan", 0))   # 0 (default value)

# 3. Remove
del scores["Rahul"]
popped = scores.pop("Ananya")   # returns 95
last_item = scores.popitem()    # removes & returns last pair (since 3.7)
```

Looping in dictonary

```python

# Only keys
for name in scores:
    print(name)

# Keys and values
for name, marks in scores.items():
    print(f"{name:8} → {marks}")

# Only values
for marks in scores.values():
    print(marks)

```

Like list comprehensions, but for dictionaries.

{key_expression: value_expression for item in iterable}

```python


# Examples
squares = {x: x**2 for x in range(1, 11)}
# {1: 1, 2: 4, 3: 9, ..., 10: 100}

# With condition
even_squares = {x: x**2 for x in range(1, 21) if x % 2 == 0}
# {2: 4, 4: 16, 6: 36, ..., 20: 400}

# From two lists (very common)
names = ["vikash", "priya", "rahul"]
scores = [88, 92, 75]
student_scores = {name: score for name, score in zip(names, scores)}
# {'vikash': 88, 'priya': 92, 'rahul': 75}
```

Dictionaries inside dictionaries — very common for structured / hierarchical data.


```python
# Example: student database
students = {
    "vikash": {
        "age": 28,
        "city": "Delhi",
        "scores": {"math": 88, "science": 92, "english": 85},
        "hobbies": ["coding", "cricket"]
    },
    "priya": {
        "age": 25,
        "city": "Mumbai",
        "scores": {"math": 95, "science": 98, "english": 90},
        "hobbies": ["reading", "dance"]
    }
}

# Accessing
print(students["vikash"]["scores"]["science"])     # 92
print(students["priya"]["hobbies"][0])             # reading

# Adding / modifying
students["vikash"]["scores"]["python"] = 94
students["rahul"] = {"age": 27, "city": "Bangalore", "scores": {}}

# Safe nested access (avoid KeyError)
score = students.get("vikash", {}).get("scores", {}).get("physics", "N/A")
```

Merge a dictonary

```python
# Old way (works everywhere)
d1 = {"a": 1, "b": 2}
d2 = {"b": 20, "c": 3}
merged = {**d1, **d2}           # {'a': 1, 'b': 20, 'c': 3}
# or
merged = d1.copy()
merged.update(d2)

# Python 3.9+  (recommended when possible)
merged = d1 | d2                # {'a': 1, 'b': 20, 'c': 3}

# Multiple merges
defaults = {"theme": "dark", "font": "Arial"}
user = {"theme": "light", "zoom": 125}
final = defaults | user | {"lang": "en"}
```
