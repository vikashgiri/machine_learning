```python
try:
    result = 10 / 0  # This will raise ZeroDivisionError
except ZeroDivisionError:
    print("Oops! Can't divide by zero, Vikash!")

print("Program continues...")  # This still runs!
```

Catch multiple exception

```python
try:
    num = 0  # Might fail if not a number
    result = 10 / num
except ValueError:
    print("That's not a number! Try again.")
except ZeroDivisionError:
    print("Zero? No way!")
```

Grouping Exceptions

```python
try:
    # Simulate file or network error
    result = 10 / 0
except (ZeroDivisionError, ValueError) as e:  # Catches both, 'as e' stores the error
    print(f"Math error: {e}")
```

else: Runs only if no exception in try. Great for "success" code.

finally: Always runs (cleanup, like closing files). Even if exception or else.

Flowchart (in words):

try → Exception? → except → finally

try → No exception → else → finally

```python
try:
    print("try executed")
except FileNotFoundError as e:
    print(f"File not found: {e}")
else:
    print("else execute successfully!")
finally:
    print("Cleanup done.")  # Always runs
    # file.close()  # Good habit, but 'with' is better (Step 6)
```
