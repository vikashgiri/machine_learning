Lambda Function: Unnamed, one expression only, no return or def. 
Limitation: Only one line/expression — no loops, if-statements (unless in a one-liner), or complex logic.
Syntex:-

lambda = keyword (starts it)

arguments = inputs (like x or x, y)

: = separator

expression = what to compute/return (single line)

```python
# Lambda to add 10 to a number
add_ten = lambda x: x + 10
print(add_ten(20))  # 30

#no argument
greet = lambda: "Hello, Vikash!"
print(greet())  # Hello, Vikash!

#multiple arg
multiply = lambda x, y: x * y
print(multiply(3, 4))  # 12
```

Use ternary: value_if_true if condition else value_if_false

```python
is_even = lambda x: "Even" if x % 2 == 0 else "Odd"
print(is_even(4))  # Even
print(is_even(7))  # Odd
```
