Functions help you organize code, reuse logic, and make things modular.

Reuse code: Write once, use many times.

Readability: Break big programs into small, named chunks.

Debugging: Easier to test and fix isolated pieces.

```python
def function_name():
    # Your code here (indented!)
    print("Hello from the function!")
```

function_name: Your choice (snake_case recommended, e.g., greet_user).

(): Empty for no inputs (parameters).

:: Starts the indented block (use 4 spaces or Tab—be consistent!).

No return? It implicitly returns None.

Quick Tip: Functions must be defined before called. Python reads top-to-bottom.

```python
def say_hello():
    print("Welcome to Python functions!")

say_hello()  # Calls it once
say_hello()  # Calls it again—reuse!
```

Parameters are placeholders in the definition; arguments are the actual values you pass when calling.

```python
def greet_user(name):  # 'name' is a parameter
    print(f"Hello, {name}!")

greet_user("Vikash")  # "Vikash" is the argument
# Output: Hello, Vikash!
```

Pass args by name—great for clarity.

```python
def introduce(name, age):
    print(f"Hi, I'm {name}, and I'm {age} years old.")

introduce(age=28, name="Vikash")  # Order doesn't matter now!
```

Set defaults for optional args (must come after required ones).


```python
def add_numbers(a, b=0):  # b defaults to 0
    result = a + b
    print(f"{a} + {b} = {result}")

add_numbers(5)       # 5 + 0 = 5
add_numbers(5, 3)    # 5 + 3 = 8
add_numbers(b=10, a=2)  # 2 + 10 = 12
```

Functions can give back results with return. Without it, they return None.

```python
def square(num):
    return num ** 2  # Exits function immediately

result = square(4)
print(result)  # 16
print(square(5))  # 25 (direct use)
```

Return multiple values as a tuple (unpacking works!).


```python
def divide(a, b):
    return a // b, a % b  # Integer division + remainder

quotient, remainder = divide(10, 3)
print(quotient, remainder)  # 3 1
```

variable Scope

Local: Inside function (e.g., x = 5 in def f():—only visible inside).

Global: Outside functions (visible everywhere, but modify carefully).

Enclosing/Nonlocal: For nested functions (advanced, next lesson)

```python
x = 10  # Global

def func():
    x = 20  # Local—shadows global
    print(f"Inside: {x}")  # 20

func()
print(f"Outside: {x}")  # Still 10 (global unchanged)
```

To modify global: Use global keyword (rarely—prefer passing as params).


```python
counter = 0

def increment():
    global counter
    counter += 1

increment()
print(counter)  # 1
```

For unknown number of args. *args = positional (tuple), **kwargs = keyword (dict) used in function.

```python
def sum_all(*args):  # args is a tuple
    return sum(args)

print(sum_all(1, 2, 3))     # 6
print(sum_all(10))          # 10
print(sum_all())            # 0 (empty tuple)
```

```python
def print_info(**kwargs):  # kwargs is a dict
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Vikash", age=28, city="Delhi")
# name: Vikash
# age: 28
# city: Delhi
```

Functions can live inside others

```python
def outer():
    def inner():  # Only visible inside outer
        print("I'm inner!")
    
    inner()  # Call it here

outer()  # Output: I'm inner!
# inner()  # Error outside!
```

Closures (Inner "Remembers" Outer Variables)

```python
def make_multiplier(factor):
    def multiplier(x):  # Inner
        return x * factor  # "Closes over" factor
    return multiplier  # Return inner as function

times_two = make_multiplier(2)
print(times_two(5))  # 10
times_three = make_multiplier(3)
print(times_three(5))  # 15
```
