Combining sets with dictionaries 

```python
data = {
    "name": "Vikash",
    "age": 28,
    "city": "Delhi",
    "temp": 25.5,
    "debug": True,
    "salary": 85000
}

keep_fields = {"name", "age", "city", "salary"}
clean_data = {k: v for k, v in data.items() if k in keep_fields}

print(clean_data)
# {'name': 'Vikash', 'age': 28, 'city': 'Delhi', 'salary': 85000}
```

combine set/dict/list in one. Very frequent pattern in data processing:


```python
# Example: Find students who passed at least one subject
records = [
    {"name": "Rahul", "subjects": ["math", "science", "english"]},
    {"name": "Priya", "subjects": ["math", "history"]},
    {"name": "Vikash", "subjects": ["science", "physics"]},
    {"name": "Ananya", "subjects": []}
]

passed_subjects = {"math", "science", "physics"}   # set for fast lookup

successful_students = [
    student["name"]
    for student in records
    if set(student["subjects"]) & passed_subjects   # intersection non-empty
]

print(successful_students)
# ['Rahul', 'Priya', 'Vikash']
```

Deduplicate list of dictionaries by a key


```python
items = [
    {"id": 101, "name": "Laptop"},
    {"id": 102, "name": "Phone"},
    {"id": 101, "name": "Laptop"},  # duplicate
    {"id": 103, "name": "Tablet"}
]

seen_ids = set()
unique_items = []

for item in items:
    if item["id"] not in seen_ids:
        unique_items.append(item)
        seen_ids.add(item["id"])

print([item["name"] for item in unique_items])
# ['Laptop', 'Phone', 'Tablet']
```
