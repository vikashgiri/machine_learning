Tuples are used to store multiple items in a single variable, just like lists — but with one huge difference: tuples are immutable.

```python
# Most common ways
t1 = (1, 2, 3)                  # with parentheses (recommended for clarity)
t2 = 10, 20, 30                 # without parentheses (tuple packing)
t3 = ("apple",)                 # single-item tuple → comma is REQUIRED!
t4 = tuple([5, 6, 7])           # from a list
t5 = ()                         # empty tuple
```

Important trap — don't forget the comma for single-item tuples:


```python
wrong = ("hello")     # This is just a string!
correct = ("hello",)  # This is a tuple with one element
```

 Accessing & Slicing (same as lists)


```python
point = (10, 20, 30)

print(point[0])       # 10
print(point[-1])      # 30 (negative index works)
print(point[1:3])     # (20, 30) → slicing returns a new tuple
```

3. You Cannot Change a Tuple (Immutability)
Python

```python
colors = ("red", "green", "blue")

# All of these will raise TypeError:
# colors[0] = "yellow"
# colors.append("purple")
# colors.pop()
# colors[1:2] = ["lime"]
```

Use tuple when:
Data should never change (coordinates, RGB colors, days of week, config values)

Returning multiple values from a function

Using as dictionary key or set element

You want slight performance gain & memory saving

You want to signal to other programmers: "this data is fixed"

Use list when:

You need to add/remove/change elements

Order matters but content will grow/shrink

Building dynamic collections

Workaround if you really need to "change" one (create a new tuple):


```python
colors = ("red", "green", "blue")
new_colors = colors[:1] + ("yellow",) + colors[2:]
# → ('red', 'yellow', 'blue')
```

Here are the  ways to loop over tuples.

```python
point = (10, 20, 30, "origin")

for item in point:
    print(item)
# Output:
# 10
# 20
# 30
# origin
```

 list of tuples (very common pattern)

```python
students = [
    ("Rahul", 85, "A"),
    ("Priya", 92, "A+"),
    ("Vikash", 78, "B+"),
    ("Ananya", 65, "C")
]

# Basic (without unpacking)
for student in students:
    print(student)          # prints whole tuple

# With unpacking → clean & readable
for name, marks, grade in students:
    print(f"{name} | Marks: {marks} | Grade: {grade}")
```

Nested tuples (looping inside tuples)


```python
nested = (
    ("Delhi", (28.6, 77.2)),
    ("Mumbai", (19.1, 72.8)),
    ("Bangalore", (12.97, 77.59))
)

for city, (lat, lon) in nested:
    print(f"{city}: Latitude {lat}, Longitude {lon}")
```

Basic Unpacking (Most Common)

```python
person = ("Vikash", 28, "Delhi")

name, age, city = person
print(name)   # Vikash
print(age)    # 28
print(city)   # Delhi

numbers = (1, 2, 3, 4, 5, 6)

first, second, *rest = numbers
print(first)     # 1
print(second)    # 2
print(rest)      # [3, 4, 5, 6]   ← list!

# More variations:
first, *middle, last = numbers
print(first, last)      # 1 6
print(middle)           # [2, 3, 4, 5]

*beginning, second_last, last = numbers
print(second_last, last)   # 5 6

# Multiple ignore
_, _, _, n1, *_ = numbers
print(n1)   # 4
```
