pip does exactly the same thing for Python code.
You write in terminal / command prompt and computer goes to the internet and downloads  library and puts it in your system. Example like install latest numpy library

pip install numpy

PyPI (Python Package Index) is the giant online library / store / warehouse of Python packages like playstore and app store where Python libraries/packages like numpy, pandas, torch, requests etc can download like Apps (WhatsApp, Instagram, etc.) from app/play store . pip command goes there automatically when you write pip install libraryName. The url for Pypi is https://pypi.org/  . PyPI has over 760,000+ different packages right now and More than 17 million files uploaded (different versions, wheels, source code).

The package will downloaded from PyPI in your local system in your system's main Python folders.
depend on your operating system. If you want to see location run this command

Lets You install package A of version 1.2. Later you install package B which use package A but with diffrent version of A like version 2.0. It will raise a conflict.So either old project breaks, or new project breaks.

The Solution is that each project gets its own private Python and own private packages will be better.
That private folder and is called virtual environment.

Instead of just typing pip install ...
Many experienced people (and official docs) now recommend:

python -m pip install numpy

#OR

python3 -m pip install numpy

#OR if you are using many python version

py -3.11 -m pip install numpy

Why? → It guarantees you're using the correct Python's pip (especially if you have Python 3.9, 3.10, 3.11, 3.12 installed together).

Imagine your laptop has one single global Python (the one you installed).
You install packages pandas lib globally using pip install .
Now look at this realistic situation:

you have a Old college project running on library pandas 1.5.x and a New AI project (2026)runing on library pandas 2.2.x (latest). When you Upgrade pandas than old project breaks!

So we need Virtual Environment.

A virtual environment creates a seprate folder for each project where all lib will live in that folder.When you activate it than pip install goes only there (global Python stays clean)
Each project gets its own isolated folder.

So Each project = own locked room (.venv) → install whatever you want inside, no effect on other projects and Avoid version conflicts (most important reason)
Different projects need different versions of the same library so venv lets both live happily on same laptop in diffrent folder like One project on Python 3.10, another on 3.12, another on 3.13 — all on same computer. if you Want to remove a project than Just delete the .venv folder — zero trace left.

We need a simple text file (usually named exactly requirements.txt) that lists:

All the Python libraries (packages) your project needs
Usually exact versions so that everyone gets the same environment

numpy==1.26.4

pandas==2.2.2

matplotlib==3.9.2


When someone gets your project folder, they just run one command and everything installs correctly:
Bash

pip install -r requirements.txt

Important note:-

Keep separate files if needed
requirements.txt for production / main app and
requirements-dev.txt for only for developers (pytest, black, jupyter, etc.)

Install dev extras: pip install -r requirements-dev.txt

Never commit huge .venv/ folder to git → add .venv/ to .gitignore

Always commit requirements.txt to git

it is technically possible to use (create and install from) a requirements.txt file without any virtual environment.
But in practice, it's usually a bad idea f

venv → classic, safe, everywhere (still #1 recommendation for learning)

uv → fastest modern replacement (very strong trend in 2025–2026)

conda/miniconda/micromamba → when you need numpy/scipy/pytorch with CUDA easily

Everything else is either legacy or niche.

For agentic AI work (building agents with LangGraph, CrewAI, LlamaIndex, AutoGen, etc.) the virtual environment choice is not the most important decision — but it still matters for speed, reliability, and avoiding version hell.

Install UV

(1)Open CMD

(2)check UV is exist by 'uv --version'

(3)If not installed install it by command powershell -ExecutionPolicy ByPass -c "irm https://astral.sh/uv/install.ps1 | iex"

(4)After it finishes, close the terminal window completely and open a new Command Prompt again. and check UV again

(5)create a new folder and create a virtual environment by command 'uv venv'

(6)Did you see .venv folder appear? (You can check in File Explorer → go to the folder)

(7)Activate it '.venv\Scripts\activate' and You should now see (.venv) at the beginning of the prompt.

(8)uv init to create a pyproject.toml

 (9)Install lib what you requied like 'uv add numpy pandas'

(10)See lib list by 'uv pip list'

When you create a project with uv, it uses two main files:


(1)pyproject.toml → where you write what you want (the "human-readable" list)

name = "my-agent"

version = "0.1.0"

requires-python = ">=3.11"

dependencies = [

    "langgraph >= 0.2",

    "langchain-openai",

    "pydantic",
]

(2)uv.lock → uv automatically creates and updates this file


uv.lock is a special file that uv creates and manages automatically.
Inside it, uv writes down Exact version of every package your project uses
(not just the ones you wrote with uv add, but also all their sub-dependencies / children / grandchildren packages).uv reads uv.lock  installs exactly these versions and Without lock file the pip install today command might pull pandas 2.2.2, tomorrow pandas 2.3.0 (breaks code) etc

Commit uv.lock to Git (yes — you should commit it!) so that Everyone pulls  and  runs 'uv sync' command to install identical lib/environment

Command 'uv sync' reads uv.lock and installs exactly library version into .venv 

NOTE- Never edit uv.lock by hand — it's machine-generated (like a JSON from npm)

Always commit uv.lock to git 

 requirements.txt is NOT required when you use uv properly.
