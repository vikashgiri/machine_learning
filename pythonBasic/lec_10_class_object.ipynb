constructor is the special method called __init__ (double underscore init).
It is automatically called right after an object is created (when you do obj = ClassName(...)).Name must be exactly __init__ and First parameter must be self and It does not return anything

Main purposes:

Initialize instance attributes

Set up the initial state of the object

Run any startup logic that every object needs



```python
class Student:
    school = "Delhi University"
    
    def __init__(self, name, roll_no, branch='cs'): #constructor
        self.name = name
        self.roll_no = roll_no
        self.branch=branch
        self.grades = []          # Instance list
    
    def add_grade(self, score):
        self.grades.append(score)
    
    def average(self):
        if self.grades:
            return sum(self.grades) / len(self.grades)
        return 0

vikash = Student("Vikash", 101)
vikash.add_grade(85)
vikash.add_grade(92)
print(f"Average: {vikash.average()}")  # 88.5
```

Inheritance?
Purpose: Reuse code from a parent class in child classes without duplicating it.
Syntax: Use parentheses after the class name to specify the parent: class Child(Parent):.
Key Idea: A child class gets all attributes (variables) and methods (functions) from the parent for free. It can then add, modify, or override them.

```python
# Parent class
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    
    def start_engine(self):
        return f"{self.brand} {self.model}'s engine started!"

# Child class inheriting from Vehicle
class Car(Vehicle):
    def __init__(self, brand, model, doors):
        super().__init__(brand, model)  # We'll explain super() later; for now, it calls parent's __init__
        self.doors = doors
    
    def honk(self):
        return "Beep beep!"

# Create an instance
my_car = Car("Toyota", "Camry", 4)
print(my_car.start_engine())  # Inherited from Vehicle
print(my_car.honk())         # New method in Car
print(f"Doors: {my_car.doors}")  # New attribute in Car
```

Quick Check: What happens if you remove super().__init__? The brand and model won't be set—always initialize parents properly!

Sometimes, you want the child to behave differently. Overriding lets you redefine a parent method in the child.

```python
class ElectricCar(Car):  # Car inherits from Vehicle, so this is multi-level
    def __init__(self, brand, model, battery_size):
        super().__init__(brand, model, 4)  # Calls Car's __init__, which calls Vehicle's
        self.battery_size = battery_size
    
    def start_engine(self):  # Overrides parent's method
        return f"{self.brand} {self.model} silently activates {self.battery_size}kWh battery!"

# Usage
tesla = ElectricCar("Tesla", "Model 3", 75)
print(tesla.start_engine())  # Overridden version
print(tesla.honk())         # Inherited from Car (not overridden)
```

super() refers to the parent class dynamically. It's crucial for:

Calling parent methods without hardcoding the class name.
Handling multiple parents (advanced).

super() creates a "method resolution order" chain: SportsCar → Car → Vehicle.
Without super(), attributes from earlier classes wouldn't be set

```python
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
        print("Vehicle initialized")

class Car(Vehicle):
    def __init__(self, brand, model, doors):
        super().__init__(brand, model)  # Calls Vehicle's __init__
        self.doors = doors
        print("Car initialized")

class SportsCar(Car):
    def __init__(self, brand, model, top_speed):
        super().__init__(brand, model, 2)  # Calls Car's __init__ (which calls Vehicle's)
        self.top_speed = top_speed
        print("SportsCar initialized")

# Usage
ferrari = SportsCar("Ferrari", "488", 200)
print(ferrari.top_speed)
```

Python supports inheriting from multiple classes. The child gets members from all parents.

```python
class Flyable:
    def fly(self):
        return "Flying high!"

class Swimmable:
    def swim(self):
        return "Swimming deep!"

class Duck(Flyable, Swimmable):  # Multiple parents
    def quack(self):
        return "Quack!"

# Usage
donald = Duck()
print(donald.fly())    # From Flyable
print(donald.swim())   # From Swimmable
print(donald.quack())  # Own method
```

When a method exists in multiple parents, Python uses MRO(Method resolution order) to decide which one to call. It's a depth-first, left-to-right traversal (C3 linearization algorithm).

View MRO with ClassName.__mro__ or ClassName.mro().
Diamond Problem: If two parents inherit from the same grandparent, MRO prevents duplicate calls.

```python
class A:  # Grandparent
    def greet(self):
        return "Hello from A"

class B(A):
    def greet(self):
        return super().greet() + " (via B)"

class C(A):
    def greet(self):
        return super().greet() + " (via C)"

class D(B, C):  # Multiple: B and C both inherit from A
    pass

d = D()
print(d.greet())
print(D.__mro__)  # Shows the order: D -> B -> C -> A -> object
```

Explanation:

MRO: D → B → C → A → object.
greet in D isn't defined, so it goes to B (first in order), which calls super() to C, then A.
Without MRO, you'd get duplicates or errors—Python's C3 solves this elegantly.

Advanced Tip: Design inheritance hierarchies to avoid deep diamonds; prefer composition over deep inheritance.

Abstract Base Class (ABC)  module use to create abstract classes. These can't be instantiated and force children to implement methods.Imagine you are making rules for a game:
Every animal must make a sound.
But you don’t want anyone to create a plain “Animal” — only real animals like Dog, Cat, Cow are allowed.
So you say:
→ “If you want to be called an Animal, you must tell me what sound you make.
→ If you don’t, I will not allow you to play (i.e. you cannot create the object).”
This is exactly what ABC + @abstractmethod does in Python.
@abstractmethod makes methods mandatory in subclasses.

```python
from abc import ABC, abstractmethod

# This is the "rule book" — no one is allowed to make a plain Animal
class Animal(ABC):

    # This is the rule: every real animal MUST have this method
    @abstractmethod
    def sound(self):
        pass   # We don't write code here — children must write it

# ----------------------

# This is allowed (follows the rule)
class Dog(Animal):
    def sound(self):           # ← must write this
        return "Woof woof!"

# This is allowed
class Cat(Animal):
    def sound(self):
        return "Meow!"

# ----------------------

# This is NOT allowed — will give error
class Robot(Animal):
    pass   # Forgot to write sound() method

# Try to create objects
d = Dog()     # OK
c = Cat()     # OK
#r = Robot()   # ERROR!
```
