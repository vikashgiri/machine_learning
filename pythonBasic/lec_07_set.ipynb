Sets in Python are  unordered collection of unique and immutable elements.
They are very useful when you want to remove duplicates,Check membership quickly (very fast — O(1) average time)
Perform mathematical set operations (union, intersection, difference, symmetric difference)

```python
# Most common ways
s1 = {1, 2, 3, 2, 4}           # duplicates removed → {1, 2, 3, 4}
s4 = set()                     # empty set (important!)
```

Basic operation on set

```python
fruits = {"apple", "banana", "cherry", "apple"}  # → {'apple', 'banana', 'cherry'}
# Add 
fruits.add("mango")
#removed
fruits.remove("banana")        # KeyError if not found
fruits.discard("kiwi")         # no error if missing
fruits.pop()                   # removes & returns arbitrary element

# Membership (very fast!)
print("apple" in fruits)       # True
print("grape" not in fruits)   # True

print(len(fruits))             # number of unique items
```

3. Set Operations (the real power)


```python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7}

# Union (all elements)
print(A | B)          # {1, 2, 3, 4, 5, 6, 7}
print(A.union(B))     # same

# Intersection (common elements)
print(A & B)          # {4, 5}
print(A.intersection(B))

# Difference (A - B: in A but not in B)
print(A - B)          # {1, 2, 3}
print(A.difference(B))

# Symmetric difference (elements in exactly one set)
print(A ^ B)          # {1, 2, 3, 6, 7}
print(A.symmetric_difference(B))

# Update in place (modifies original)
A |= {8, 9}           # A becomes {1,2,3,4,5,8,9}
A &= B                # A becomes intersection
```

Set Comprehension


```python
# Like list/dict comprehension
squares = {x**2 for x in range(10)}          # {0, 1, 4, 9, 16, 25, 36, 49, 64, 81}

evens = {x for x in range(20) if x % 2 == 0} # {0, 2, 4, ..., 18}
```

Very Common Real-World Patterns


```python
# 1. Remove duplicates from list
numbers = [5, 2, 8, 2, 1, 5, 9]
unique = list(set(numbers))          # [5, 2, 8, 1, 9] (order not guaranteed)

# 2. Find common / different items
friends_A = {"Rahul", "Priya", "Vikash", "Ananya"}
friends_B = {"Vikash", "Amit", "Priya", "Neha"}

common = friends_A & friends_B               # mutual friends
only_A   = friends_A - friends_B             # only in A
either   = friends_A ^ friends_B             # exclusive friends

# 3. Fast membership check (much faster than list)
allowed = {"read", "write", "execute"}
if "admin" in allowed:
    print("Access granted")
```
